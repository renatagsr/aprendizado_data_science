# Data Storytelling

<center><img width = '45%' src = 'https://thumbor.forbes.com/thumbor/711x711/https://blogs-images.forbes.com/brentdykes/files/2016/03/data_story3_sq-1.jpg?width=960'></center>

Storytelling é a capacidade de contar bem as histórias, pode ser aplicado em diversas áreas. Para um cientista de dados é de grande importância que se utilize storytelling para que possa contar de forma clara, objetiva e bonita a linha de pensamento para obter os resultados desejados. Outra importância para a sua utilização é porque o visual vende, ou seja, o visual tem a capacidade de convencimento e melhora a capacidade de entendimento/engajamento.

Nesse notebook será apresentado um exemplo de como é diferente a apresentação de resultados de forma padrão para uma apresentação com storytelling. 

```
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
```

Forma de plotar os dados agrupados.

```
labels = ['Fortemente\nContra', 'Contra', 'Neutro', 'Favorável', 'Fortemente\nFavorável']
x = np.arange(len(labels))
width_bar = 0.35
antes = [22, 37, 17, 15, 9]
depois = [10, 10, 11, 30, 39]
```

```
fig, ax = plt.subplots(figsize = (8, 4))

retangulo_1 = ax.bar(x, antes, label = "Antes")
retangulo_2 = ax.bar(x, depois, label = "Depois")

ax.set_xticks(x)
ax.set_xticklabels(labels)

plt.tight_layout()
plt.plot()
```

Dessa forma é possível plotar as barras do gráfico de forma desagrupada e ainda ter uma separação entre elas.

```
fig, ax = plt.subplots(figsize = (8, 4))

rects1 = ax.bar(x - width_bar/2 - 0.05, antes, width_bar, label = "Antes")
rects2 = ax.bar(x + width_bar/2, depois, width_bar, label = "Depois")

ax.set_xticks(x)
ax.set_xticklabels(labels)

plt.tight_layout()
plt.plot()
```

Plotagem do gráfico em barras separadas.

```
fig, ax = plt.subplots(figsize = (8, 4))

rects1 = ax.bar(x - width_bar/2, antes, width_bar, label = "Antes")
rects2 = ax.bar(x + width_bar/2, depois, width_bar, label = "Depois")

ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_ylabel('Porcentagem')
ax.set_title('Apoio popular operação XYZ')
ax.legend()

plt.tight_layout()
plt.plot()
```

```
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```
#configurar as fontes
plt.rcParams['font.family'] = 'sans-serif'

# definindo cores do gráfico
CINZA1, CINZA2, CINZA3= "#CFCDCE", "#A2A1A3", "#787878"
AZUL1, AZUL2 = "#8db7eb", "#253760"

# definir cores por plot
cores_antes = [CINZA1, CINZA1, CINZA1, CINZA3, CINZA3]
cores_depois = [AZUL1, AZUL1, AZUL1, AZUL2, AZUL2]
```

```
#Criar fig e axes
fig, ax = plt.subplots(figsize = (8, 4))

#plotar barra antes e depois
rects1 = ax.bar(x - width_bar/2, antes, width_bar, label = "Antes", color = cores_antes)
rects2 = ax.bar(x + width_bar/2, depois, width_bar, label = "Depois", color = cores_depois)

#Labels do eixo x
ax.set_xticks(x)
ax.set_xticklabels(labels)
#ax.set_ylabel('Porcentagem')
#ax.set_title('Apoio popular operação XYZ')
#ax.legend()

#Remover bordas
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)

#Remover labels do eixo y
plt.gca().axes.get_yaxis().set_visible(False)

#Remover xticks do eixo x
plt.tick_params(
    axis = 'both',
    which = 'both',
    bottom = False,
    top = False,
    labelbottom = True
)


#Criar uma função para dar labels automaticamente
def autolabel(rects):
  for rect in rects:
        height = rect.get_height()
        ax.annotate(
                    '{}%'.format(height),
                    xy=(rect.get_x() + rect.get_width()/2,  height),
                    xytext=(0, -20),  # offset
                    textcoords="offset points",
                    ha='center',
                    va='bottom',
                    fontsize=10,
                    color="white",
                    )

#Plotar labels
autolabel(rects1)
autolabel(rects2)

#Título
plt.text(-0.35, 52, 
         'Apoio popular à Operação XYZ',
         fontsize = 16,
         color = CINZA3, weight = 'bold')

#Subtítulo
plt.text(-0.35, 49, 
         'Efetividdade da campanha\n'
         'junto a população ribeirnha',
         color = CINZA2,
         va ='top')

# plotar anotações do resultado da campanha
plt.text(4.5, 39,
         "$\\bf{RESULTADO}$\n"
         "de uma campanha\n"
         "de marketing junto\n"
         "à população, onde o\n"
         "apoio chegou a 39%."
         ,
         verticalalignment="top",
         fontsize=12,
         color=AZUL2)

plt.tight_layout()
plt.plot()
```
