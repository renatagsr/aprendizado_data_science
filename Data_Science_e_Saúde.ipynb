```
deaths = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
recovered = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')

display(deaths)
display(recovered)
```

```
def latest_by_country(data):
  return data.groupby('Country/Region').sum().iloc[:,-1]
display(latest_by_country(confirmed).head())
display(latest_by_country(deaths).head())
display(latest_by_country(recovered).head())
```

```
informations = [latest_by_country(confirmed), latest_by_country(deaths), latest_by_country(recovered)]
combined = pd.concat(informations, axis = 1)
combined.columns = ['Confirmed', 'Deaths', 'Recovered']
combined.head()
```

Taxa de letalidade é expressa em % o número de casos letais dentro dos casos que foram letais e não letais.

Imagina que existe uma doença que em janeiroa de 2020 afeou 100 pessoas, 5 pessoas morreram (por exemplo), letalidade é de 5%. Não digo que ela é de 5% para sempre porque a doença continuou existindo.

**Conclusão**: é MUITO difícil de falar de letalidade tentando ser preciso durante uma epidemia. No começo é comum que somente casos graves sejam reportados.

```
sum_up = combined.sum()
letality_rate_1 = sum_up['Deaths']/sum_up['Confirmed']
letality_rate_2 = sum_up['Deaths']/(sum_up['Recovered'] + sum_up['Deaths'])
print(letality_rate_1 * 100)
print(letality_rate_2 * 100)
```

A diferença de um para outro provavelmente se deve a demora de recuperação. Era esperado que os valores fossem iguais

Isso mostra que, provavelmente, a segunda letalidade é uma métrica pessimista.

```
def latest_by_country_at(data, date):
  return data.groupby('Country/Region').sum().iloc[:,-1]

informations = [latest_by_country_at(confirmed, '9/6/20'), latest_by_country_at(deaths, '9/6/20'), latest_by_country_at(recovered, '9/6/20')]
combined_9_6_20 = pd.concat(informations, axis = 1)
combined_9_6_20.columns = ['Confirmed', 'Deaths', 'Recovered']
combined_9_6_20.head()

sum_up = combined_9_6_20.loc['China']
letality_rate_1 = sum_up['Deaths']/sum_up['Confirmed']
letality_rate_2 = sum_up['Deaths']/(sum_up['Recovered'] + sum_up['Deaths'])
print(letality_rate_1 * 100)
print(letality_rate_2 * 100)
```

Nesse caso está considerando um período de 12 dias, desde quando a pessoa foi diagnosticada até ela ser curada ou falecer. 

O resultado está dando igual pois, como pode ser visto no gráfico, as taxas estão se mantendo praticamente constantes desde março/abril.

```
informations = [latest_by_country_at(confirmed, '8/27/20'), latest_by_country_at(deaths, '9/8/20'), latest_by_country_at(recovered, '9/8/20')]
combined_12 = pd.concat(informations, axis = 1)
combined_12.columns = ['Confirmed', 'Deaths', 'Recovered']
combined_12.head()

sum_up = combined_12.loc['China']
letality_rate_3 = sum_up['Deaths']/sum_up['Confirmed']
print(letality_rate_3 * 100)
```

```
letality_rate_1 = combined['Deaths']/combined['Confirmed'] * 100
letality_rate_2 = combined['Deaths']/(combined['Recovered'] + combined['Deaths']) * 100
combined['letality_rate_1'] = letality_rate_1
combined['letality_rate_2'] = letality_rate_2
combined.head()
```

```
combined.sort_values('letality_rate_1', ascending=False).head(10)
```

Por estar pegando os dados do último dia, mostra que realmente a segunda métrica de letalidade é pessimista, porque a soma dos casos de óbito e de casos recuperados não são iguais. Isso implica que realmente a recuperação é lenta.

É muito ruim e muito difícil fazer comparação com períodos de tempo diferentes. Quer extrapolar simplesmente e multipicar por 6? A mortalidade será muito alta, mas extrapolar com esse tipo de crescimento pode ser muito irreal. Então o jeito é ir por estimativas... Estatística.

```
grouped_per_day = pd.DataFrame([confirmed.sum()[3:-1], recovered.sum()[2:-1], deaths.sum()[2:-1]],
             index= ['Confirmed', 'Recovered', 'Deaths']).T
grouped_per_day.head()
```

```
grouped_per_day['letality_rate_1'] = grouped_per_day['Deaths']/grouped_per_day['Confirmed'] * 100
grouped_per_day['letality_rate_2'] = grouped_per_day['Deaths']/(grouped_per_day['Deaths'] + grouped_per_day['Recovered']) * 100
grouped_per_day.head()
```

```
plt.figure(figsize=(10,5))
grouped_per_day['letality_rate_1'].plot()
ax = grouped_per_day['letality_rate_2'].plot()
plt.legend(['Mortes/Casos confirmados', 'Mortes/(Mortes + Recuperados'])
plt.title('Letalidade "pessimista e otimista" confirmados no mundo todo, convergindo para o mesmo valor')
plt.grid(ls = '-')
ax.set_xlabel('Fonte: curso www.alura.com.br e dados Johns Hopkins https://github.com/CSSEGISandData/COVID-19/tree/master/csse_covid_19_data')
ax.set_ylabel('% Letalidade')

plt.show()
```
