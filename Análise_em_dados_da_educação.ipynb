```
provas_entra = ['NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_CN', 'NU_NOTA_REDACAO']
prova_saida = 'NU_NOTA_MT'
dados_sem_notas = dados_sem_notas_zero[provas].dropna()
notas_entrada = dados_sem_notas[provas_entra]
notas_saida = dados_sem_notas[prova_saida]

notas_entrada
```

Faz a retirada de tudo que tem "NaN" no conjunto de dados de provas para que seja possível fazer o modelo de machine learning.

```
x = notas_entrada
y = notas_saida
```

Em "x" está os valores que será fornecido para treinar o modelo, o "y" são os valores que quer na saída. É como se fosse uma função onde o x são os valores e o y o resultado ao utilizar os valores de x dentro da função.

```
from sklearn.model_selection import train_test_split
```

```
SEED = 4321
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.25, random_state = SEED)
```

Coloca um valor fixo para o "random_state" porque caso não se coloque a cada vez que rodar os dados de treino e de teste, irão gerar valores diferentes, colocando um valor fixo ainda irá gerar valores aleatórios, porém de uma forma fixa.

Os dados x e y tem que gerar a mesma quantidade de respostas.

```
from sklearn.svm import LinearSVR
```

```
modelo = LinearSVR(random_state = SEED)
modelo.fit(x_treino, y_treino)
```

```
predicoes_matematica = modelo.predict(x_teste)
```

Esse gráfico mostra os erros. Percebe-se que os erros foram tanto para baixo quanto para cima.
```
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.25)
modelo_arvore = DecisionTreeRegressor(max_depth = 3)
modelo_arvore.fit(x_treino, y_treino)
predicoes_matematica_arvore = modelo_arvore.predict(x_teste)
mean_squared_error(y_teste, predicoes_matematica_arvore)
```

A aleatoriedade do modelo não está só no tipo de teste realizado, mas também na forma que os dados foram divididos.

```
from sklearn.model_selection import cross_validate
```

```
modelo_arvore = DecisionTreeRegressor(max_depth = 2)
resultados = cross_validate(modelo_arvore, x, y, cv = 10, scoring = 'neg_mean_squared_error')
media = (resultados['test_score']*(-1)).mean()
media
```

```
modelo_arvore = DecisionTreeRegressor(max_depth = 2)
resultados = cross_validate(modelo_arvore, x, y, cv = 10, scoring = 'neg_mean_squared_error')
media = (resultados['test_score']*(-1)).mean()
desvio_padrao = (resultados['test_score']*-1).std()
lim_inferior = media - (2 * desvio_padrao)
lim_superior = media + (2 * desvio_padrao)
print(f'O intervalo de confiança é {lim_inferior} - {lim_superior}')
```

Coloca o "scoring" para pode analisar igual estava sendo analisando antes, quando utilizava a outra métrica.

```
resultados['test_score']*-1
```

```
def calcula_mse(resultados):
  media = (resultados['test_score']*(-1)).mean()
  desvio_padrao = (resultados['test_score']*-1).std()
  lim_inferior = media - (2 * desvio_padrao)
  lim_superior = media + (2 * desvio_padrao)
  print(f'O intervalo de confiança é {lim_inferior} - {lim_superior}')
```

```
from sklearn.model_selection import cross_validate
from sklearn.model_selection import KFold
import numpy as np

SEED = 1232
np.random.seed(SEED)

partes = KFold(n_splits = 10, shuffle = True)
modelo_arvore = DecisionTreeRegressor(max_depth = 2)
resultados = cross_validate(modelo_arvore, x, y, cv = partes, scoring = 'neg_mean_squared_error')
calcula_mse(resultados)
```
